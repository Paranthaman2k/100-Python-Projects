1. and --> condtion1 and condition2

```python
marks = int(input("enter your marks: "))

if marks>=80:
  print("Distinction with PASS")
elif marks >= 60 and marks < 80:
  print("First Class")
elif marks >= 36 and marks < 60:
  print("Second class")
elif marks == 35:
  print("Just PASS")
else:
  print("Fail")
```
